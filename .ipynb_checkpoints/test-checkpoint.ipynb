```python
from MinPy.reg import AutoReg
import torch as t
import numpy as np
from MinPy.net import dmf
import MinPy.train.data_loader as data_loader
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

n = 100

I_n = t.from_numpy(np.eye(n)).to(t.float32)
#print(L_model)
W_model = dmf.net([n,n],bias_if=False,nonlinear=False)
W = W_model(I_n)

W = data_loader.get_data(pic_type='real',pic_name='./train_pics/double_Cameraman.jpg',height=n,width=n)
plt.imshow(W,'gray')
plt.show()
DE_model = AutoReg.net(n)
optimizer = t.optim.SGD(DE_model.parameters(),lr=0.01)

for epoch in range(20001):
    DE = DE_model(W,mode='row')
    optimizer.zero_grad()
    DE.backward(retain_graph=True)
    optimizer.step()
    if epoch%1000 == 0:
        print('Epoch=',epoch,',Dirac Energy:',DE.data)
    if epoch%5000 == 0:
        print(DE_model.A_4.data)
        sns.heatmap(DE_model.A_4.data*(DE_model.A_4.data<-0.3))
        plt.show()







```
